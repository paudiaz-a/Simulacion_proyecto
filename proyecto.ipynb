```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. TUS DATOS REALES (Frecuencias extradas de las imgenes) ---
# Definimos los bins (Lmite superior) hasta 45 min como pediste
bins_labels = ['15-20', '20-25', '25-30', '30-35', '35-40', '40-45']
x = np.arange(len(bins_labels))  # Posiciones en el eje X

# Frecuencias por horario (Datos extrados de tus imgenes hasta el corte de 45)
# H1: 6-9
h1_freq = [1, 9, 39, 55, 32, 15] 
# H2: 9-14
h2_freq = [0, 5, 28, 43, 21, 4]
# H3: 14-18
h3_freq = [1, 7, 11, 31, 14, 6]
# H4: 18-22
h4_freq = [8, 30, 70, 28, 2, 1]

datos_lista = [h1_freq, h2_freq, h3_freq, h4_freq]
titulos = ["09:00 - 14:00","06:00 - 09:00", "14:00 - 18:00", "18:00 - 22:15"]

# --- 2. CONFIGURACIN DEL PANEL 2x2 ---
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Distribucin de Frecuencia de Salidas por Horario', fontsize=16, fontweight='bold')
axs = axs.flatten()

# --- 3. GENERACIN DE GRFICOS ---
for i, ax in enumerate(axs):
    frecuencias = datos_lista[i]
    
    # Barras
    barras = ax.bar(x, frecuencias, color='#4c72b0', edgecolor='black', alpha=0.8, width=0.6)
    
    # Lnea de Tendencia (Polgono)
    ax.plot(x, frecuencias, color='#c44e52', marker='o', linewidth=2, label='Tendencia')

    # Lneas de Meta (Objetivos 10 y 15 min)
    # Nota: Como nuestros bins empiezan en 15, las metas estn "fuera a la izquierda"
    # Dibujamos una flecha indicando que la meta est antes


    # Etiquetas y Formato
    ax.set_title(titulos[i], fontsize=12, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(bins_labels, rotation=45)
    ax.set_ylabel('Frecuencia (Viajes)')
    ax.set_xlabel('Minutos entre Salidas')
    ax.grid(axis='y', linestyle='--', alpha=0.3)
    
    # Mostrar valor encima de cada barra
    for rect in barras:
        height = rect.get_height()
        if height > 0:
            ax.text(rect.get_x() + rect.get_width()/2., height + 0.5,
                    f'{int(height)}', ha='center', va='bottom', fontsize=9)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. TUS DATOS REALES (Frecuencias extradas de las imgenes) ---
# Definimos los bins con sus puntos medios reales para espaciado proporcional
bins_labels = ['15-20', '20-25', '25-30', '30-35', '35-40', '40-45']
# Puntos medios de cada intervalo para espaciado proporcional
x = np.array([17.5, 22.5, 27.5, 32.5, 37.5, 42.5])
# Ancho de cada bin (todos son de 5 minutos)
width = 4  # Ancho visual de las barras (ajustable)

# Frecuencias por horario (Datos extrados de tus imgenes hasta el corte de 45)
# H1: 6-9
h1_freq = [1, 9, 39, 55, 32, 15] 
# H2: 9-14
h2_freq = [0, 5, 28, 43, 21, 4]
# H3: 14-18
h3_freq = [1, 7, 11, 31, 14, 6]
# H4: 18-22
h4_freq = [8, 30, 70, 28, 2, 1]

datos_lista = [h1_freq, h2_freq, h3_freq, h4_freq]
titulos = ["09:00 - 14:00","06:00 - 09:00", "14:00 - 18:00", "18:00 - 22:15"]

# --- 2. CONFIGURACIN DEL PANEL 2x2 ---
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Distribucin de Frecuencia de Salidas por Horario', fontsize=16, fontweight='bold')
axs = axs.flatten()

# --- 3. GENERACIN DE GRFICOS ---
for i, ax in enumerate(axs):
    frecuencias = datos_lista[i]
    
    # Barras con espaciado proporcional
    barras = ax.bar(x, frecuencias, width=width, color='#4c72b0', 
                   edgecolor='black', alpha=0.8, align='center')
    
    # Lnea de Tendencia (Polgono)
    ax.plot(x, frecuencias, color='#c44e52', marker='o', linewidth=2, label='Tendencia')

    # Etiquetas y Formato
    ax.set_title(titulos[i], fontsize=12, fontweight='bold')
    
    # Configurar etiquetas del eje X para mostrar lmites continuos
    x_ticks = [15, 20, 25, 30, 35, 40, 45]
    ax.set_xticks(x_ticks)
    ax.set_xticklabels([str(tick) for tick in x_ticks])
    
    ax.set_ylabel('Frecuencia (Viajes)')
    ax.set_xlabel('Minutos entre Salidas')
    ax.grid(axis='y', linestyle='--', alpha=0.3)
    
    # Configurar lmites del eje X para mejor visualizacin
    ax.set_xlim(15, 45)
    
    # Mostrar valor encima de cada barra
    for rect in barras:
        height = rect.get_height()
        if height > 0:
            ax.text(rect.get_x() + rect.get_width()/2., height + 0.5,
                    f'{int(height)}', ha='center', va='bottom', fontsize=9)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
